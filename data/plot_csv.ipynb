```python
# Loaded variable 'df' from URI: /vlt/devel/graphics/textualize/projects/sparkline/data/short.csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

df = pd.read_csv(r'/vlt/devel/graphics/textualize/projects/sparkline/data/long.csv')
df.set_index('second', inplace=True)
# df.columns = ['seconds', 'speed']
# df.to_csv(r'/vlt/devel/graphics/textualize/projects/sparkline/data/short.csv')
```

```python
plt.plot(df)
plt.xlabel(df.index.name)
plt.ylabel(df.columns[0])
```

```python
df.shape
```

```python
noise = pd.DataFrame(np.random.default_rng().exponential(scale=1.0, size=df.shape[0]), index=df.index)
noise.columns = ['speed']
noise
```

```python

df_noise_upper = df + noise
# df_noise
df_noise_upper.columns = ['upper']
plt.plot(df_noise_upper)
plt.xlabel(df_noise_upper.index.name)
plt.ylabel(df_noise_upper.columns[0])
```

```python

noise = pd.DataFrame(np.random.default_rng().exponential(scale=1.0, size=df.shape[0]), index=df.index)
noise.columns = ['speed']
df_noise_lower = df - noise
# df_noise
df_noise_lower.columns = ['lower']
```

```python

df2 = pd.concat([df, df_noise_upper, df_noise_lower], axis='columns')
# df2
plt.plot(df2)
df2.to_csv(r'/vlt/devel/graphics/textualize/projects/sparkline/data/long_noise.csv')
```
